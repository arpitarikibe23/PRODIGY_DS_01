```python
df = pd.read_csv("/content/drive/MyDrive/PRODIGY_DS_01/titanic/test.csv")
df = pd.read_csv("/content/drive/MyDrive/PRODIGY_DS_01/titanic/train.csv")
```

**Histogram for Age Distribution**

```python
# Step 4: Plot a histogram for the 'Age' column
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**Bar Chart for Gender Distribution**

```python
# Step 5: Plot a bar chart for the 'Sex' column
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex', palette='pastel')
plt.title('Gender Distribution of Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Bar Chart for Passenger Class Distribution**

```python
# Step 6: Plot a bar chart for Passenger Class ('Pclass')
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass', palette='muted')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
